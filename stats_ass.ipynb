```python
categorical_columns = ['Date', 'Day',	'SKU',	'City',	'Volume',	'BU',	'Brand',	'Model'] # Example, replace with actual categorical column names

for col in categorical_columns:
    if col in df.columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, x=col)
        plt.title(f'Frequency of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        # Print basic value counts for insights
        print(f"\nDistribution of {col}:")
        print(df[col].value_counts())

        # Provide some basic insights
        print(f"\nInsights for {col}:")
        most_common = df[col].value_counts().idxmax()
        least_common = df[col].value_counts().idxmin()
        print(f"- The most frequent category is '{most_common}'.")
        print(f"- The least frequent category is '{least_common}'.")
        print(f"- There are {df[col].nunique()} unique categories.")
        if df[col].isnull().any():
            print(f"- There are {df[col].isnull().sum()} missing values in this column.")
    else:
        print(f"Column '{col}' not found in the DataFrame.")
```

```python

numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

# Create a new DataFrame to store the standardized data to avoid modifying the original
df_standardized = df.copy()

# Before standardization: Plot original distributions
print("Distributions Before Standardization:")
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Original Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Apply standardization
for col in numerical_columns:
    mean = df[col].mean()
    std = df[col].std()
    # Avoid division by zero if standard deviation is 0
    if std > 0:
        df_standardized[col] = (df[col] - mean) / std
    else:
        df_standardized[col] = df[col] - mean # If std is 0, all values are the same, centering around 0

# After standardization: Plot standardized distributions
print("\nDistributions After Standardization:")
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_standardized[col], kde=True)
    plt.title(f'Standardized Distribution of {col}')
    plt.xlabel(f'Standardized {col}')
    plt.ylabel('Frequency')
    plt.show()

# Display descriptive statistics after standardization
print("\nDescriptive Statistics After Standardization:")
print(df_standardized[numerical_columns].describe())
```

```python

categorical_cols_for_ohe = ['Day',	'SKU',	'City',	'BU',	'Brand',	'Model']

# Apply one-hot encoding to the selected categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_cols_for_ohe, drop_first=False) # drop_first=True avoids multicollinearity

# Display the first few rows of the new dataframe with one-hot encoded columns
print("DataFrame after One-Hot Encoding:")
print(df_encoded.head())

# Display the shape of the new dataframe to see the increase in columns
print("\nShape of the DataFrame after One-Hot Encoding:", df_encoded.shape)
```

```python

print("\nNew columns created by One-Hot Encoding:")
# We can find the original categorical columns that were encoded
original_categorical_cols = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# The new columns will be the original columns plus the new one-hot encoded columns
# excluding the original categorical columns that were replaced.
# We can identify the new columns by comparing the columns before and after encoding.

# Get columns before encoding (excluding columns that were replaced)
cols_before_encoding = df.columns.drop(original_categorical_cols)
# Get columns after encoding
cols_after_encoding = df_encoded.columns

# Find the columns that are in the encoded dataframe but not in the original dataframe
new_encoded_columns = cols_after_encoding.difference(cols_before_encoding)

print(new_encoded_columns)

# Display the first few rows again to show the binary values
print("\nSample of the DataFrame showing binary encoded columns:")
# Select a few of the new encoded columns to display
sample_encoded_cols = new_encoded_columns[:10] # Take the first 10 new columns as an example
print(df_encoded[sample_encoded_cols].head())
```
